```python
import pandas as pd
from pandas_datareader import data as pdr
from datetime import datetime
import yfinance as yfin #important libriary, without it DataReader can't get data because wrong type
yfin.pdr_override()

stocks_list = ["META",
               "AMZN",
               "NFLX",
               "GOOG"]

start = datetime(datetime.now().year-1,
                 datetime.now().month,
                 datetime.now().day)

end = datetime.now()

for stock in stocks_list:
 globals()[stock] = pdr.get_data_yahoo(stock,start,end)
```

```python
moving_average_intervals = [5, 20, 50]

for moving_average in moving_average_intervals:

  column_name = "moving_average for %s days" %(str(moving_average))

  GOOG[column_name] = GOOG["Adj Close"].rolling(moving_average).mean() #rolling function will make moving average due to column that we select
  #now we will have 3 new columns: moving average for 5, 20 and 50 days

GOOG[["Adj Close",
      "moving_average for 5 days",
      "moving_average for 20 days",
      "moving_average for 50 days"]].plot(figsize=(14,6),
                                          title="Moving average")
```

```python
GOOG["Daily Return"] = GOOG["Adj Close"].pct_change()

GOOG["Daily Return"].plot(figsize=(14,14),
                          legend=True,
                          linestyle="--",
                          marker = 'o')
```

```python
adjusted_closing_dataframe = pdr.get_data_yahoo(stocks_list,start,end)['Adj Close']
print(adjusted_closing_dataframe)
```
